#### 1.6. Formato Long - Wide

A menudo, cuando trabajamos con datos, es 칰til organizarlos de manera que podamos entenderlos y analizarlos m치s f치cilmente. Uno de los conceptos importantes en este proceso es el formato "Long" y "Wide" en un DataFrame.

**Formato Long**:
- En el formato "Long", los datos se organizan de manera vertical, con una columna para las categor칤as y otra para los valores. Cada fila representa una observaci칩n 칰nica y puede incluir un identificador 칰nico para esa observaci칩n.
- Este formato es 칰til cuando tienes una gran cantidad de categor칤as y deseas una representaci칩n m치s compacta de tus datos.

**Formato Wide**:
- En el formato "Wide", los datos se organizan de manera horizontal, con una fila para cada observaci칩n y una columna para cada categor칤a. Cada celda en la tabla contiene un valor 칰nico.
- Este formato es 칰til cuando deseas una vista m치s amplia y clara de tus datos, especialmente cuando tienes un n칰mero limitado de categor칤as.

**Transformaci칩n entre formatos**:
- A menudo, necesitas transformar tus datos de un formato a otro para realizar an치lisis espec칤ficos. Esto se puede lograr utilizando herramientas de programaci칩n como Pandas en Python.
- La transformaci칩n entre formatos Long y Wide implica reorganizar tus datos de una forma a la otra, manteniendo la integridad de la informaci칩n.

```python
# Probablemente te estar치s preguntando "쯔 qu칠 se referir치 con <<en caso de ser factible>>?". Resulta que no siempre ser치 posible crear bucles para tratar nuestras bases de datos, ya que esto depende en gran medida de la estructura de los datos con los que trabajemos. Si tenemos una estructura de datos que lo permita, porque por ejemplo trabajamos con un libro de Excel y los formatos entre hojas son equivalentes, podemos crear un bucle para el mismo tratamiento que aplicamos a una hoja. En caso de ser diferentes, esto no podr치 ser as칤, y tendremos que adecuarnos a la estructura de datos de cada una. Pero no se preocupen, esta herramienta en verdad es f치cilmente aplicable, especialmente si quieren consolidar registros de una misma fuente pero en diferentes fechas, por ejemplo.

#Pasos
#Crear una lista con los nombres de las hojas
sheets = ['1' , '2']

#creo una lista vacia donde se van a almacenar las bases de datos para cada mes una vez termine el tratamiento
dfs = []

for sheet in sheets:
    # Leer hoja y saltar las primeras 8 filas
    df = pd.read_excel('ipc-julago2023.xlsx', sheet_name=sheet, skiprows=8)

    # Eliminar columna 'A침o corrido'
    df.drop(['A침o corrido'], axis=1, inplace=True)

    # Eliminar filas con valores nulos
    df.dropna(inplace=True)

   # Renombrar columnas
    df.rename(columns={'Unnamed: 0': 'Fecha'}, inplace=True)
    df.rename(columns={'Mensual': 'Inflaci칩n Mensual'}, inplace=True)
    df.rename(columns={'Anual': 'Inflaci칩n Anual'}, inplace=True)

    # Agregar DataFrame a la lista
    dfs.append(df)

#concateno los resultados y los guardo en mi lista vacia
df_final = pd.concat(dfs, ignore_index=True)
df_final
```
